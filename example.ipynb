```python
from paptools import Number, Array
import sympy as sp


# Example usage of paptools

# Creating a Number with uncertainty, a symbolic expression, and units
a = Number(5.0, 0.2, symbol='a', unit='m')
b = Number(3.0, 0.1, symbol='b', unit='m')

# Performing arithmetic operations
c = a + b

print("Print the result with propagated uncertainty and units:")

print(c)
print("\n")


print("Print the symbolic expression and its error propagation:")

print(sp.latex(c.get_expr()))
print(sp.latex(c.get_err_expr()))
print("\n")


print("Render the result in LaTeX format in Jupyter Notebook:")
c
```

```python
# Create an Array

x = Array([1.1, 2.3, 3.4], [0.1, 0.2, 0.1], symbol='x', unit='kg') # Array with different uncertainties
y = Array([4.2, 5.2, 6.5], 0.1, symbol='y', unit='m')  # Array with same uncertainty for all elements
z = Array([7.3, 8.1, 9.6], symbol='z', unit='s**2') # Array without uncertainties

c = x * y / z

#Convert the Unit to milli Newtons
c.convert_to_unit('mN')


print("Print the result with propagated uncertainty and units:")

print(c)
print("\n")

print("Print the rounded result with propagated uncertainty and units:")
c.round()
print(c)
print("\n")


print("Print the symbolic expression and its error propagation:")

print(sp.latex(c.get_expr()))
print(sp.latex(c.get_err_expr()))
print("\n")


print("Render the result in LaTeX format in Jupyter Notebook:")
c
```
